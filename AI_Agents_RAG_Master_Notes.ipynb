<div class="chapter-banner">🚪 CHAPTER 1 &nbsp;|&nbsp; Introduction to AI Agents & RAG</div>

<div class="dlbook"><div class="glass-card">
<h3>🤖 What is an AI Agent?</h3>
<p>An <b>AI Agent</b> is an LLM-powered system that can <b>reason, plan, use tools, and act</b>
autonomously to achieve a goal — going beyond a single prompt-response into multi-step workflows.</p>
<h3>📚 What is RAG?</h3>
<p><b>Retrieval-Augmented Generation (RAG)</b> grounds an LLM's answers in external knowledge by
retrieving relevant documents at query time, instead of relying purely on the model's frozen training data.</p>
<table>
<tr><th>Aspect</th><th>Plain LLM</th><th>RAG</th><th>AI Agent</th></tr>
<tr><td>Knowledge</td><td>Frozen at training time</td><td>Live external documents</td><td>Live + Tools + Memory</td></tr>
<tr><td>Action</td><td>Text only</td><td>Text only</td><td>Can call APIs/tools</td></tr>
<tr><td>Autonomy</td><td>None</td><td>None</td><td>Multi-step planning</td></tr>
<tr><td>Example</td><td>ChatGPT base answer</td><td>Chat with your PDF</td><td>Auto research assistant</td></tr>
</table>
</div></div>

<div class="chapter-banner">🧩 CHAPTER 2 &nbsp;|&nbsp; LLM Foundations Recap</div>

<div class="dlbook"><div class="glass-card">
<h3>Tokenization</h3><p>Text is split into sub-word tokens (e.g. Byte-Pair Encoding) before being fed to the model.</p>
<h3>Embeddings</h3><p>Each token/sentence is mapped to a dense vector capturing semantic meaning — the backbone of retrieval and similarity search.</p>
<h3>Context Window</h3><p>The maximum number of tokens an LLM can process at once — a key constraint for RAG and agent design.</p>
</div></div>

<div class="dlbook"><div class="formula-box">similarity(A,B) = (A · B) / (‖A‖‖B‖) &nbsp;&nbsp;— Cosine Similarity between embedding vectors</div></div>

```python
# Tokenization + Embedding + Cosine similarity demo
sentences = ["AI agents can use tools", "Agents combine reasoning and action", "The weather today is sunny"]
vocab = sorted(set(w.lower() for s in sentences for w in s.split()))
vec = {w:i for i,w in enumerate(vocab)}

def bow_embed(sentence, dim=len(vocab)):
    v = np.zeros(dim)
    for w in sentence.lower().split():
        v[vec[w]] += 1
    return v / (np.linalg.norm(v)+1e-8)

embs = np.array([bow_embed(s) for s in sentences])
def cosine(a,b): return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b)+1e-8)

print("Similarity (sent1, sent2):", round(cosine(embs[0],embs[1]),3))
print("Similarity (sent1, sent3):", round(cosine(embs[0],embs[2]),3))
```

<div class="dlbook"><div class="summary-box"><b>📦 Chapter Summary</b><ul><li>Tokenization converts text to model-readable sub-word units.</li><li>Embeddings place semantically similar text close together in vector space.</li><li>Cosine similarity is the standard metric for comparing embedding vectors.</li></ul></div></div>

<div class="dlbook"><div class="formula-box">ANN Search: find top-k vectors v ∈ V minimizing distance(query, v) — approximate for speed at scale</div></div>

```python
# Simple k-NN vector search from scratch
np.random.seed(0)
doc_embeddings = np.random.randn(8,4)
doc_texts = [f"Document {i}" for i in range(8)]
query = np.random.randn(4)

def cosine_sim(a,b): return (a@b)/(np.linalg.norm(a)*np.linalg.norm(b)+1e-8)
sims = [cosine_sim(query, d) for d in doc_embeddings]
top_k = np.argsort(sims)[::-1][:3]

print("Query vector:", np.round(query,2))
print("\nTop-3 retrieved documents:")
for idx in top_k:
    print(f"  {doc_texts[idx]} | similarity = {sims[idx]:.3f}")

plt.figure(figsize=(7,3.5))
plt.bar(doc_texts, sims, color=[NEON[0] if i in top_k else '#3a3f7a' for i in range(8)])
plt.title("Vector Search — Similarity Scores (highlighted = retrieved)", color='#e6f1ff')
plt.xticks(rotation=30); plt.ylabel("Cosine Similarity")
plt.show()
```

<div class="chapter-banner">📖 CHAPTER 4 &nbsp;|&nbsp; Retrieval-Augmented Generation (RAG) Deep Dive</div>

<div class="dlbook"><div class="glass-card">
<h3>RAG Pipeline</h3>
<ol>
<li><b>Ingest:</b> Load documents (PDFs, web pages, etc.)</li>
<li><b>Chunk:</b> Split into smaller passages</li>
<li><b>Embed:</b> Convert chunks into vectors</li>
<li><b>Store:</b> Save vectors in a vector database</li>
<li><b>Retrieve:</b> Find top-k relevant chunks for a query</li>
<li><b>Augment + Generate:</b> Feed retrieved chunks + query into the LLM prompt</li>
</ol>
</div></div>

<div class="dlbook"><div class="formula-box">Prompt = SystemInstructions + Retrieved_Context(top-k chunks) + UserQuery &nbsp;→&nbsp; LLM → Grounded Answer</div></div>

```python
# RAG pipeline flow diagram
fig, ax = plt.subplots(figsize=(10,3))
stages = ["Documents","Chunking","Embedding","Vector DB","Retrieve\nTop-K","LLM +\nContext","Answer"]
xs = np.linspace(0,12,len(stages))
for i,(x,s) in enumerate(zip(xs,stages)):
    ax.scatter(x,0, s=1000, color=NEON[i%len(NEON)], zorder=3, edgecolor='white')
    ax.text(x,0.7, s, ha='center', color='#e6f1ff', fontsize=9, fontweight='bold')
    if i < len(stages)-1:
        ax.annotate("", xy=(xs[i+1]-0.55,0), xytext=(x+0.55,0), arrowprops=dict(arrowstyle="->", color="#00f0ff", lw=2))
ax.set_xlim(-1,13); ax.set_ylim(-1,1.3); ax.axis('off')
ax.set_title("End-to-End RAG Pipeline", color='#e6f1ff')
plt.show()
```

```python
# Minimal RAG implementation from scratch (NumPy only)
docs = [
    "AI agents combine reasoning with tool use to complete tasks.",
    "RAG retrieves relevant documents before generating an answer.",
    "Vector databases store embeddings for fast similarity search.",
    "LangChain is a popular framework for building LLM applications.",
]
def fake_embed(text, dim=16, seed_offset=0):
    rng = np.random.RandomState(abs(hash(text)) % (2**31))
    return rng.randn(dim)

doc_vecs = np.array([fake_embed(d) for d in docs])
query = "How do AI agents use tools?"
q_vec = fake_embed(query)

sims = doc_vecs @ q_vec / (np.linalg.norm(doc_vecs,axis=1)*np.linalg.norm(q_vec)+1e-8)
top_idx = np.argsort(sims)[::-1][:2]
retrieved = [docs[i] for i in top_idx]

prompt = "Context:\n" + "\n".join(retrieved) + f"\n\nQuestion: {query}\nAnswer using only the context above."
print("🔍 Retrieved chunks:")
for r in retrieved: print(" -", r)
print("\n📝 Final Prompt sent to LLM:\n", prompt)
```

<div class="chapter-banner">🧠 CHAPTER 2 &nbsp;|&nbsp; LLM Foundations Recap</div>

<div class="dlbook"><div class="glass-card">
<p>Before building agents, recall the LLM core building blocks:</p>
<ul>
<li><b>Tokenization:</b> splitting text into sub-word tokens.</li>
<li><b>Embeddings:</b> dense vectors capturing semantic meaning — the backbone of retrieval.</li>
<li><b>Context Window:</b> the max tokens an LLM can process at once (prompt + history + retrieved docs).</li>
<li><b>Autoregressive Generation:</b> predicting the next token repeatedly to form a response.</li>
</ul>
</div></div>

<div class="dlbook"><div class="formula-box">similarity(A,B) = (A · B) / (‖A‖‖B‖) &nbsp;&nbsp;— Cosine similarity between embedding vectors</div></div>

```python
# Toy embedding similarity demo
np.random.seed(1)
vocab = ["cat","dog","car","truck","python","king"]
embeddings = {w: np.random.randn(8) for w in vocab}
# nudge similar concepts closer
embeddings["dog"] = embeddings["cat"] + 0.3*np.random.randn(8)
embeddings["truck"] = embeddings["car"] + 0.3*np.random.randn(8)

def cosine_sim(a,b): return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

sims = np.zeros((len(vocab),len(vocab)))
for i,w1 in enumerate(vocab):
    for j,w2 in enumerate(vocab):
        sims[i,j] = cosine_sim(embeddings[w1], embeddings[w2])

plt.figure(figsize=(5.5,4.5))
plt.imshow(sims, cmap='plasma')
plt.xticks(range(len(vocab)), vocab, rotation=45)
plt.yticks(range(len(vocab)), vocab)
plt.colorbar(label="Cosine Similarity")
plt.title("Word Embedding Similarity Matrix", color='#e6f1ff')
plt.show()
```

<div class="dlbook"><div class="summary-box"><b>📦 Chapter Summary</b><ul><li>Embeddings turn text into vectors where semantic similarity = geometric closeness.</li><li>Cosine similarity is the standard metric for comparing embeddings.</li><li>Context window size limits how much retrieved content an LLM can use at once.</li></ul></div></div>

<div class="chapter-banner">🔍 CHAPTER 4 &nbsp;|&nbsp; Retrieval-Augmented Generation (RAG) Deep Dive</div>

<div class="dlbook"><div class="glass-card">
<h3>The RAG Pipeline</h3>
<ol>
<li><b>Load:</b> ingest documents (PDF, web, docs).</li>
<li><b>Chunk:</b> split into smaller passages (e.g. 500 tokens, with overlap).</li>
<li><b>Embed:</b> convert chunks into vectors.</li>
<li><b>Store:</b> save vectors in a vector database.</li>
<li><b>Retrieve:</b> find top-k relevant chunks for a query.</li>
<li><b>Augment:</b> insert retrieved chunks into the LLM prompt.</li>
<li><b>Generate:</b> LLM produces a grounded answer.</li>
</ol>
</div></div>

<div class="dlbook"><div class="formula-box">Answer = LLM( Prompt_Template( Query, Top_k_Retrieved_Chunks ) )</div></div>

```python
# RAG pipeline flow diagram
fig, ax = plt.subplots(figsize=(10,3.5))
steps = ["📄 Load", "✂️ Chunk", "🧬 Embed", "🗄️ Store\n(Vector DB)", "🔍 Retrieve", "➕ Augment\nPrompt", "🤖 Generate"]
xs = np.linspace(0,12,len(steps))
for i,(x,s) in enumerate(zip(xs,steps)):
    ax.add_patch(plt.Rectangle((x-0.7,-0.5),1.4,1, color=NEON[i%len(NEON)], alpha=0.25, ec=NEON[i%len(NEON)], lw=2))
    ax.text(x,0,s, ha='center', va='center', color='white', fontsize=8, fontweight='bold')
    if i < len(steps)-1:
        ax.annotate("", xy=(xs[i+1]-0.75,0), xytext=(x+0.75,0), arrowprops=dict(arrowstyle="->", color="#00f0ff", lw=2))
ax.set_xlim(-1.5,13); ax.set_ylim(-1.2,1.2); ax.axis('off')
ax.set_title("End-to-End RAG Pipeline", color='#e6f1ff')
plt.show()
```

```python
# Minimal RAG implementation in pure NumPy (no external deps needed)
documents = [
    "The Eiffel Tower is located in Paris, France.",
    "Python is a popular programming language for AI.",
    "The mitochondria is the powerhouse of the cell.",
    "LangChain helps developers build LLM-powered applications.",
    "RAG combines retrieval with generation to reduce hallucination."
]

def fake_embed(text, dim=16, seed_offset=0):
    rng = np.random.RandomState(abs(hash(text)) % (2**31))
    return rng.randn(dim)

doc_embeddings = np.array([fake_embed(d) for d in documents])

def retrieve(query, k=2):
    q_emb = fake_embed(query)
    sims = doc_embeddings @ q_emb / (np.linalg.norm(doc_embeddings,axis=1)*np.linalg.norm(q_emb))
    top_k = np.argsort(-sims)[:k]
    return [(documents[i], sims[i]) for i in top_k]

query = "What does RAG do?"
results = retrieve(query, k=2)
print(f"Query: {query}\n")
for doc, score in results:
    print(f"  [score={score:.3f}] {doc}")

prompt = f"Answer using ONLY this context:\n{chr(10).join(d for d,_ in results)}\n\nQuestion: {query}"
print("\n--- Final Augmented Prompt sent to LLM ---\n")
print(prompt)
```

<div class="chapter-banner">✍️ CHAPTER 5 &nbsp;|&nbsp; Prompt Engineering for Agents</div>

<div class="dlbook"><div class="glass-card">
<ul>
<li><b>Zero-shot:</b> direct instruction, no examples.</li>
<li><b>Few-shot:</b> provide example input/output pairs.</li>
<li><b>Chain-of-Thought (CoT):</b> ask the model to reason step-by-step.</li>
<li><b>ReAct prompting:</b> interleave Reasoning and Acting — "Thought → Action → Observation" loop.</li>
<li><b>System prompts:</b> define persona, constraints, and tool access for agents.</li>
</ul>
</div></div>

<div class="dlbook"><div class="formula-box">ReAct loop: Thought<sub>t</sub> → Action<sub>t</sub> → Observation<sub>t</sub> → Thought<sub>t+1</sub> → ... → Final Answer</div></div>

```python
# Example ReAct-style trace (text simulation)
react_trace = [
    ("Thought", "I need to find the current population of Pakistan."),
    ("Action", "search('current population of Pakistan')"),
    ("Observation", "Search result: ~241 million (2024 estimate)."),
    ("Thought", "I now have enough information to answer."),
    ("Final Answer", "As of recent estimates, Pakistan's population is about 241 million."),
]
for role, text in react_trace:
    print(f"[{role}] {text}")
```

<div class="chapter-banner">🧵 CHAPTER 8 &nbsp;|&nbsp; Memory in AI Agents</div>

<div class="dlbook"><div class="glass-card">
<table>
<tr><th>Memory Type</th><th>Description</th><th>Example</th></tr>
<tr><td>Short-term</td><td>Conversation buffer within context window</td><td>Last N messages</td></tr>
<tr><td>Long-term</td><td>Persisted across sessions, often vector-based</td><td>User preferences store</td></tr>
<tr><td>Episodic</td><td>Specific past events/interactions</td><td>"Last time you asked about X..."</td></tr>
<tr><td>Semantic</td><td>General facts/knowledge learned</td><td>Domain knowledge base</td></tr>
</table>
</div></div>

```python
# Simple long-term memory using a vector store pattern
class SimpleMemory:
    def __init__(self): self.store = []
    def add(self, text, embed_fn):
        self.store.append((text, embed_fn(text)))
    def recall(self, query, embed_fn, k=2):
        q = embed_fn(query)
        sims = [(t, np.dot(e,q)/(np.linalg.norm(e)*np.linalg.norm(q))) for t,e in self.store]
        return sorted(sims, key=lambda x:-x[1])[:k]

def embed_fn(text):
    rng = np.random.RandomState(abs(hash(text))%(2**31)); return rng.randn(16)

mem = SimpleMemory()
mem.add("User prefers concise answers.", embed_fn)
mem.add("User is learning AI Agents and RAG.", embed_fn)
mem.add("User's favorite language is Python.", embed_fn)

recalled = mem.recall("What does the user like to code in?", embed_fn)
for text, score in recalled:
    print(f"[{score:.3f}] {text}")
```

<div class="dlbook"><div class="summary-box"><b>📦 Chapter Summary</b><ul><li>Short-term memory = context window; long-term memory = persisted external store.</li><li>Vector-based memory enables semantic recall, not just recency-based recall.</li><li>Good memory design prevents context bloat while preserving relevant history.</li></ul></div></div>

<div class="chapter-banner">🛠️ CHAPTER 9 &nbsp;|&nbsp; Tool Use & Function Calling</div>

<div class="dlbook"><div class="glass-card">
<p>Modern LLMs support <b>function calling</b>: the model outputs a structured JSON call to a tool
(e.g. a calculator, search API, database query) which your code executes, then feeds the result back.</p>
</div></div>

<div class="dlbook"><div class="formula-box">LLM(prompt + tool_schemas) → {"tool":"get_weather","args":{"city":"Karachi"}} → execute → feed result back → LLM final answer</div></div>

```python
# Function calling simulation
def get_weather(city):
    fake_db = {"Karachi": "34°C, sunny", "London": "15°C, rainy"}
    return fake_db.get(city, "Unknown city")

def calculator(expression):
    try: return str(eval(expression, {"__builtins__":{}}))
    except Exception as e: return f"Error: {e}"

tools = {"get_weather": get_weather, "calculator": calculator}

# Simulated model tool-call decision
tool_call = {"tool": "get_weather", "args": {"city": "Karachi"}}
result = tools[tool_call["tool"]](**tool_call["args"])
print("Tool called:", tool_call)
print("Tool result:", result)
print("Final LLM response: 'The weather in Karachi is currently", result, "'")
```

```python
# OpenAI-style function/tool schema example
tool_schema = {
    "type": "function",
    "function": {
        "name": "get_weather",
        "description": "Get current weather for a city",
        "parameters": {
            "type": "object",
            "properties": {"city": {"type": "string", "description": "City name"}},
            "required": ["city"]
        }
    }
}
import json as _json
print(_json.dumps(tool_schema, indent=2))
```

<div class="dlbook"><div class="summary-box"><b>📦 Chapter Summary</b><ul><li>Function calling lets LLMs trigger real code execution via structured output.</li><li>Tool schemas describe name, description, and parameters so the model calls correctly.</li><li>Always validate/sandbox tool arguments before executing — never blindly eval user input in production.</li></ul></div></div>

<div class="dlbook"><div class="glass-card">
<ul>
<li><b>Chain-of-Thought:</b> linear step-by-step reasoning, no tool use.</li>
<li><b>ReAct:</b> interleaves reasoning with tool actions (Ch.5).</li>

</ul>

```python
# Plan-and-Execute simulation
goal = "Write a 3-bullet market summary for an AI startup"
plan = [
    "Step 1: Search recent AI market trends",
    "Step 2: Identify top 3 relevant trends",
    "Step 3: Draft 3 concise bullet points",
    "Step 4: Self-review for clarity and correctness",
]
print(f"GOAL: {goal}\n")
print("Generated Plan:")
for step in plan: print(" -", step)

print("\nExecuting...")
results = {
    "Step 1": "Found trends: Agentic AI, RAG adoption, smaller efficient models.",
    "Step 2": "Top 3: Agentic AI, RAG, efficient small models.",
    "Step 3": "• Agentic AI is rising\n• RAG is becoming standard\n• Smaller efficient models are trending",
    "Step 4": "Reviewed: bullets are clear and concise. ✅ Approved.",
}
for k,v in results.items():
    print(f"\n[{k}] -> {v}")
```

<div class="chapter-banner">🏗️ CHAPTER 12 &nbsp;|&nbsp; Building a RAG Pipeline from Scratch</div>

<div class="dlbook"><div class="glass-card"><p>Let's assemble everything from Chapters 3-4 into one complete, runnable mini-RAG system.</p></div></div>

```python
# Complete mini RAG system (pure Python/NumPy, framework-free)
class MiniRAG:
    def __init__(self, dim=16):
        self.dim = dim
        self.docs = []
        self.embeddings = None

    def _embed(self, text):
        rng = np.random.RandomState(abs(hash(text)) % (2**31))
        return rng.randn(self.dim)

    def add_documents(self, texts, chunk_size=50):
        for t in texts:
            words = t.split()
            for i in range(0, len(words), chunk_size):
                self.docs.append(" ".join(words[i:i+chunk_size]))
        self.embeddings = np.array([self._embed(d) for d in self.docs])

    def retrieve(self, query, k=3):
        q = self._embed(query)
        sims = self.embeddings @ q / (np.linalg.norm(self.embeddings,axis=1)*np.linalg.norm(q))
        idx = np.argsort(-sims)[:k]
        return [(self.docs[i], sims[i]) for i in idx]

    def generate(self, query, k=3):
        context = self.retrieve(query, k)
        context_text = "\n".join(f"- {d}" for d,_ in context)
        prompt = f"Context:\n{context_text}\n\nQuestion: {query}\nAnswer based only on the context above."
        return prompt  # in production: pass this prompt to an actual LLM call

rag = MiniRAG()
rag.add_documents([
    "RAG stands for Retrieval-Augmented Generation, combining a retriever with an LLM.",
    "Vector databases like FAISS, Chroma and Pinecone store embeddings for fast similarity search.",
    "Agents extend LLMs with memory, tools, and iterative reasoning loops like ReAct.",
])

final_prompt = rag.generate("What is RAG?")
print(final_prompt)
```

<div class="chapter-banner">🤖 CHAPTER 13 &nbsp;|&nbsp; Building an AI Agent from Scratch</div>

<div class="dlbook"><div class="glass-card"><p>A minimal agent loop: observe → think → act → observe result → repeat until done.</p></div></div>

```python
# Minimal agent loop implementation (rule-based simulation of an LLM's tool decisions)
def calculator(expr):
    try: return str(eval(expr, {"__builtins__":{}}))
    except Exception as e: return f"Error: {e}"

def search(query):
    fake_results = {"capital of France": "Paris", "speed of light": "299,792 km/s"}
    return fake_results.get(query, "No result found.")

TOOLS = {"calculator": calculator, "search": search}

def fake_llm_decide(task, step):
    # Simulates an LLM deciding the next action (deterministic for demo purposes)
    if "calculate" in task.lower() and step == 0:
        return {"thought": "This requires math, I'll use the calculator.",
                "action": "calculator", "args": task.split("calculate")[-1].strip()}
    elif "capital" in task.lower() and step == 0:
        return {"thought": "This requires a lookup, I'll search.",
                "action": "search", "args": "capital of France"}
    return {"thought": "I have enough information.", "action": "final_answer", "args": None}

def run_agent(task, max_steps=4):
    print(f"TASK: {task}\n")
    observation = None
    for step in range(max_steps):
        decision = fake_llm_decide(task, step)
        print(f"Step {step}: Thought -> {decision['thought']}")
        if decision["action"] == "final_answer":
            print(f"Final Answer -> {observation}")
            break
        observation = TOOLS[decision["action"]](decision["args"])
        print(f"  Action -> {decision['action']}({decision['args']!r})  =>  Observation: {observation}\n")

run_agent("calculate 25 * 4 + 10")
print("\n" + "="*50 + "\n")
run_agent("what is the capital of France")
```

<div class="dlbook"><div class="summary-box"><b>📦 Chapter Summary</b><ul><li>The core agent loop is: Think → Act (tool call) → Observe → repeat → Final Answer.</li><li>Real agents use the LLM itself (not hardcoded rules) to decide each action.</li><li>Adding a max_steps / termination condition prevents infinite loops.</li></ul></div></div>

<div class="dlbook"><div class="glass-card">
<ul>
<li><b>Caching:</b> cache embeddings & frequent queries to cut cost/latency.</li>
<li><b>Rate limiting & cost control:</b> guard against runaway agent loops calling expensive APIs.</li>

</ul>
</div></div>
